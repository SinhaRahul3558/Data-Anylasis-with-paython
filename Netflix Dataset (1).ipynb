# Top 10 Countries on Netflix

```python
df.head(2)
```

```python
df['first_country'] = df['Country'].apply(lambda x: x.split(",")[0])
```

```python
countries = df['first_country'].value_counts()
countries
```

```python
countries = df['first_country'].value_counts().rename_axis('Country')
countries
```

```python
countries = df['first_country'].value_counts().rename_axis('Country').reset_index(name='counts')
countries
```

```python
countries = df['first_country'].value_counts().rename_axis('Country').reset_index(name='counts').head(10)
countries
```

```python
plt.figure(figsize = (15,7))
sns.barplot(x=countries['Country'].head(10),y=countries['counts'],edgecolor="black")
plt.title("Top 10 Countries on Netflix", fontfamily='serif',fontweight='bold',fontsize=20)
plt.show()
```
